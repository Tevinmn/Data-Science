# User defined functions

# Function definition

In python we use the word *def* to define a function, followed by round brackets and a colon.

The colon is used to indicate the start of a new block of code.

# Function name

This is the name that will be used later to call the function.

The name must **begin with a letter or an underscore, must be in lowercase, can have numbers** and **can be any length**

The name **cannot** be the same as any **Python keyword**, and mus describe the **function's purpose** for readability and ease of understanding.

# Function inputs/parameters

They define the **variables required for the function to perform its tasks**.

Inputs are not mandatory. If the function tasks do not require any inputs, we can leave them as empty.


### Difference between a parameters and arguments

**parameters** are variables listed in the function declaration, serving as palceholders for the vaues that the function will reveive during its execution.

**arguments** - The actual values passed to a function when it is called. These values are asisgned to the corresponding parameters defined in the function, allowing for dynamic and flexible behaviour.

# Function instructions/tasks
This is the block of code within the function that serves as the **set of isntructions** that the function must follow.

To indicate that the code that follows belong to the function, we always indent.

# Return statement
It defines what the result of the function or the output should be.

A return statement is not required for the fucntion to eb created successfully.

However, because every function in Python must return something, functions that do not have a return statement return None by default.


### Example
Creating a function to calculate the area of a rectangle based on inputs of a user.

```python
def rectangle_area():
  length = float(input("Enter the length of the rectangle:"))

  width = float(input("Enter the width of the rectangle:"))

  area = length * width

  return f"Area:{area} cm^2"

rectangle_area()

```

# Doc Strings

They are used to **describe functionality** of a function, including the arguments and expected return, if they exist

They can also be used in other code constructs such as classes and methods, to provide aditional information to the user.

A docsttring is most commonly defined under the function definition with triple double quotation marks and is divided into 3 distinct sections:

- Functionality of the function.

- Breakdown of the parameters, if applicable.

- Breakdown of the output, if applicable.


Docstrings are not required for function to run. however, they make collaboration easier by **improving readability** and **explainability**.


To print a docstring, you can use the syntax: *help(function_name)*

```python

def area(length, width):
  """
  Calculates the area of a given rectangle by multiplying its lenght and width

  Parameters
  ----------
  length: The length of the rectangle

  width: The width of the rectangle

  Return
  ----------
  A string representing the area of the rectangle
  """

  area = length * width

  return f"Area:{area}"

area(20,10)
```

# Scope

It consists of two types of varaibles: **local variable** and **global variable**

| Feature  | Global Variable | Local Variable |
|----------|----------------|----------------|
| Location | Outside functions | Inside functions |
| Scope | Whole program | Only inside function |
| Lifetime | Entire program execution | Function execution only |
| Access | Anywhere | Only within function |

# Exercise 2

Write a function that checks whether a number is even or odd.

```python
def check_even_odd(number):
  if(number%2==0):
    print(f"{number} is even")

  else:
    print(f"{number} is odd")

check_even_odd(2)
check_even_odd(11)

```

# Exercise 3:
Write a function that takes a list of numbers and returns the sum of only the even numbers

```python
numbers = [10,21, 35, 46, 52, 66, 77, 99, 81, 100]


def sum_even_numbers(numbers):
  sum = 0
  for number in numbers:
    if(number % 2 == 0):
      sum+=number

  return sum

result = sum_even_numbers(numbers)

print(result)
```
