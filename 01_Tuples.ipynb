# Creating a tuple

A tuple is created by **assigning a group of comma-separated values** enclosed within **round brackets** to a variable.

```python
animal_tuple = ("Aika", "Mammal", "Dog", "Chihuahua")
print(animal_tuple)
```

We can also create a tuple using the **tuple constructor.**

```python
animal_tuple = tuple(["Pretty","Mammal","Puppy","Bulldog", 1, 10.5])

print(animal_tuple)
```

# Tuple Type

To verify that the variable is a tuple, we can pass the variable name to the type() function

```python
print(type(animal_tuple))
```

# Indexing

- Since tuples are ordered, each element is assigned an index that corresponds to its position, starting with index 0.
- To return the index of a specified value, we can use the **index()** method.

```python
print(animal_tuple[1])

print(f"Index of Name:{animal_tuple.index("Pretty")}")

print(f"Index of Age:{animal_tuple.index(1)}")
```

# Accessing Elements at Specific Positions using their Indices

```python
print(f"Name:{animal_tuple[0]}")
```

# Slicing
- It allows one to return a portion of tuple instead of just a single value.
- They syntax is *tuple[start:end]* with the start index being included and the end index is excluded.

```python
print(animal_tuple[1:3])
```

# Modification
- Tuples are immutable, meaning we cannot change, add, or remove items from a tuple once it has been created.

# Concatenation
- We can add elements to an existing tuple by appending another tuple to it using the + operator

# Task
- In the animal tuple, add a colour value(white).

```python
animal_tuple  = animal_tuple + ("White",)
print(animal_tuple)
```

# Duplicates

Since tuples are ordered, they **allow for duplicates** as they can be differentiated by their index.

```python
vehicle_type = ("Car", "Bicycle", "Car", "Matatu", "Bus", "Motorbike", "Scooter", "Train")

print(vehicle_type)
```

# Determine the Length of  a Tuple
- We use the len() method

```python
print(f"Length of Animal Tuple:{len(animal_tuple)}")

print(f"Length of the Vehicle Tuple:{len(vehicle_type)}")
```

# Counting the number of Occurrences of a Particular Element

```python
print(f"Number of times car appears:{animal_tuple.count("Car")}")

print(f"Number of times car appears:{vehicle_type.count("Car")}")
```

# Tuple Unpacking
- It is the assignment of tuple elements to individual variables that can be used separatley

```python
person = ("John", 20, "Male","78", "Langata", "Strathmore")

print(type(person))

name,age, gender, weight, residence, school = person

print(f"Name:{name}\n"
      f"Age:{age}\n"
      f"Gender:{gender}\n"
      f"Weight:{weight}\n"
      f"Residence:{residence}\n"
      f"School:{school}")
```
